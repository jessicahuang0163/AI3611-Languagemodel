```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ffnn = np.array([256.3, 253.47, 252.43, 255.14])
rnn = np.array([184.08, 165.76, 161.69, 159.92])
lstm = np.array([131.53, 120.05, 113.94, 114.83])
transformer = np.array([155.46, 143.45, 143.50, 134.47])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
index = np.arange(4)
bar_width = 0.2
ax.bar(index - 1.5 * bar_width, ffnn - 100, bar_width, bottom=100, label='FFNN', color='#6B92E8')
ax.bar(index - 0.5 * bar_width, rnn - 100, bar_width, bottom=100, label='RNN', color='#F0B090')
ax.bar(index + 0.5 * bar_width, lstm - 100, bar_width, bottom=100, label='LSTM', color='#9DF022')
ax.bar(index + 1.5 * bar_width, transformer - 100, bar_width, bottom=100, label='Transformer', color='#D8C2F2')
ax.set_xticks(index)
ax.set_xticklabels(['Small', 'Medium', 'Full', 'Full_variant'])
ax.set_ylabel('Test PPL')
ax.set_xlabel('Scale of parameters')
ax.legend()
plt.savefig('view_performance.pdf', dpi=400, bbox_inches='tight')
```
